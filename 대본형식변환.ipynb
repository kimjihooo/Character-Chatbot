```python
import pandas as pd
df = pd.read_csv(
    '/content/drive/MyDrive/데이터/인물 챗봇/응답하라 1988 _ 제 09 화 _ 선을 넘는다는 것.txt',
     sep="/n",header=None)
df.head()
```

```python
# 괄호안 문자 제거
import re
def del_text(text):
    return re.sub('\([^)]+\)', '', text)
```

```python
df[0] = df[0].apply(del_text)
```

```python
df['Q'] = df[0].shift(1).fillna(0)
df['A'] = df[0]
df
```

```python
df.drop(columns = 0, axis =1, inplace=True)
```

```python
df['label'] = df['A'].str[:3]
```

```python
df
```

```python
df['label'].value_counts()
```

택\t\t, 선우\t, 정환\t, 동룡\t

```python
df = df[(df['label']=='택\t\t') | (df['label']=='선우\t') | (df['label']=='정환\t') | (df['label']=='동룡\t')]
```

```python
df
```

```python
df['Q'] = df['Q'].str[3:]
```

```python
df['A'] = df['A'].str[3:]
```

```python
df
```

```python
df.dropna(axis=0, inplace = True)
```

```python
def alpha_num(text):
  return re.sub(r'[^가-힣ㄱ-ㅎㅏ-ㅣa-zA-Z0-9 ]', '', text)
```

```python
df['Q'] = df['Q'].apply(alpha_num)
df['A'] = df['A'].apply(alpha_num)
df['label'] = df['label'].apply(alpha_num)
```

```python
df
```

```python
import numpy as np
df.replace('', np.nan, inplace=True)
```

```python
df.dropna(axis=0, inplace = True)
```
